```python
df = df.replace('?' , np.NaN)
```

```python
df=df.dropna()
```

```python
Gender = df['gender']
Gender = Gender.map({'Male' : 1 , 'Female' : 0})
df['gender']= Gender
```

```python
df
```

```python
target = df['income']
```

```python
target
```

```python
dataset = df[['age' , 'fnlwgt' , 'educational-num' ,'hours-per-week' , 'gender']]
```

```python
dataset
```

```python
targt = []
for i in target:
    if i == '<=50K':
        targt.append(1)
    else:
        targt.append(0)
targt
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dataset, targt, random_state=1)
```

```python
from sklearn import naive_bayes as nb
accuracies = []
scores = []
```

```python
classifier = nb.GaussianNB()
#classifier=ensemble.RandomForestClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
# y_predict = classifier.predict(X_test)
# from sklearn.metrics import accuracy_score
# accuracy_score(y_test, y_predict)
```

```python
from sklearn.metrics import f1_score
predicted = classifier.fit(X_train, y_train).predict(X_test)
score = f1_score(y_test, predicted, average = 'weighted')
scores.append(score)
incorrect = (y_test != predicted).sum()
accuracy = (len(X_test) - incorrect) / len(X_test) * 100.
accuracies.append(accuracy)
```

```python
print("Maximum accuracy attained ", max(accuracies))
```

```python
print("f1score  ", scores[np.argmax(accuracies)])
```
